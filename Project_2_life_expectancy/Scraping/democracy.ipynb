### Scraping democracy level

```python
# Necessary imports
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
import requests
```

```python
pd.set_option('display.max_rows', 200)
```

```python
#wiki website 
url = 'https://en.wikipedia.org/wiki/Democracy_Index'
response = requests.get(url)
```

```python
response.status_code #200 = success!
```

```python
response.text[:1000]  #First 1000 characters of the HTML
```

```python
page = response.text
```

```python
soup = BeautifulSoup(page, "lxml")
```

```python
print (soup.prettify()[:2000])
```

```python
#looking for a table 
table = soup.find(class_='wikitable')
print(table)
```

```python
#getting header
th = table.find_all('th')
for i in th:
    print (i.text)
```

```python
#getting all the  rows of the table
table_rows = table.find_all('tr')
```

```python
# creating list of lists (aka list of rows)
table = []

for tr in table_rows:
    td = tr.find_all('td')
    row = [i.text.strip() for i in td]
    table.append(row)
    print(row)   #regions will have to be removed!!!
```

```python
#convirting into a DataFrame
age = pd.DataFrame(table)
age.head(200)
```

```python
#some of the rows needs to be shifted horizontally
rows = [8,17,37, 44,56,61,72,76,90,98,103,115,123,130,133,135,138,142,160]
age.iloc[rows] = age.iloc[rows].shift (axis = 1)
age
```
