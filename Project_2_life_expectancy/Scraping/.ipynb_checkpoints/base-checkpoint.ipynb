### Scraping countries life expectancy

```python
# Necessary imports
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
import requests
```

```python
#wiki website 
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_life_expectancy' 
response = requests.get(url)
```

```python
response.status_code #200 = success!
```

```python
response.text[:1000]  #First 1000 characters of the HTML
```

```python
page = response.text
```

```python
soup = BeautifulSoup(page, "lxml")
```

```python
print (soup.prettify()[:2000])
```

```python
#looking for a table 
table = soup.find('table')
print(table)
```

```python
#getting header
th = table.find_all('th')
for i in th:
    print (i.text)
```

```python
#getting all the  rows of the table
table_rows = table.find_all('tr')
```

```python
# creating list of lists (aka list of rows)
table = []

for tr in table_rows:
    td = tr.find_all('td')
    row = [i.text.strip() for i in td]
    table.append(row)
    print(row)   #regions will have to be removed!!!
```

```python
#convirting into a DataFrame
age = pd.DataFrame(table)
age.head()
```

```python
#removing two rows and two columns
age = age.iloc[2:,:3]
```

```python
age.columns = ['rank', 'country', 'age']
```

```python
age.head(10)
```

```python
age.isnull().sum()
```

```python
#looks like we have strings as rank and age 
age.dtypes
```

```python
age.info()
```

```python
#convirting strings into numeric 
age['rank'] = pd.to_numeric(age['rank'], errors='coerce')
age.dropna(inplace=True)
age['rank'] = age['rank'].astype(int)
age['age'] = age['age'].apply(float)
age.dtypes
```

```python
age.shape
```

```python
age.dropna(inplace=True)
age.reset_index(inplace=True, drop=True)
```
