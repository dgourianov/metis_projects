### Scraping health expenditure per capita


```python
# Necessary imports
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
import requests
```

```python
#wiki website 
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_total_health_expenditure_per_capita'
response = requests.get(url)
```

```python
response.status_code #200 = success!
```

```python
response.text[:1000]  #First 1000 characters of the HTML
```

```python
page = response.text
```

```python
soup = BeautifulSoup(page, "lxml")
```

```python
print (soup.prettify()[:2000])
```

```python
#looking for a table 
table = soup.find(id='WHO')
print(table)
```

```python
#getting header
th = table.find_all('th')
for i in th:
    print (i.text)
```

```python
#getting all the  rows of the table
table_rows = table.find_all('tr')
```

```python
# creating list of lists (aka list of rows)
table = []

for tr in table_rows:
    td = tr.find_all('td')
    row = [i.text.strip() for i in td]
    table.append(row)
    print(row)   
```

```python
#convirting into a DataFrame
age = pd.DataFrame(table)
age.head()
```

```python
#removing a row and keeping only data for 2015

age = age.iloc[1:,np.r_[0,4]]
```

```python
age.columns = [ 'country', 'health_exp_per_cap']
```
