### Scraping gdp per capita

```python
# Necessary imports
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
import requests
```

```python
#wiki website 
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)_per_capita' 
response = requests.get(url)
```

```python
response.status_code #200 = success!
```

```python
response.text[:1000]  #First 1000 characters of the HTML
```

```python
page = response.text
```

```python
soup = BeautifulSoup(page, "lxml")
```

```python
print (soup.prettify()[:2000])
```

```python
#looking for a table w/needed data
table = soup.find(class_="wikitable") 

print(table)
```

```python
#getting header
th = table.find_all('th')
for i in th:
    print (i.text)
```

```python
#getting all the  rows of the table
table_rows = table.find_all('tr')
```

```python
# creating list of lists (aka list of rows)
table = []

for tr in table_rows:
    td = tr.find_all('td')
    row = [i.text.strip() for i in td]
    table.append(row)
    print(row)   
```

```python
#convirting into a DataFrame
age = pd.DataFrame(table)
age.head()
```

```python
#removing two rows and two columns
age = age.iloc[1:,:3]
```

```python
age.columns = ['rank', 'country', 'gdp_cap']
```
